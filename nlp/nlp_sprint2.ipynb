# 4. Tratamento com a lib Spacy

## 4.1. Setup do spaCy

```python
spacy.cli.download('pt_core_news_sm')

spacy.cli.download('pt_core_news_lg')
```

```python
nlp = spacy.load("pt_core_news_lg")
nlp
```

## 4.2 Tirar as stop words

```python
# Definir a função que você deseja aplicar
def stop_words(valor):
    doc = nlp(valor)
    tokens_sem_stopwords = [token.text for token in doc if not token.is_stop]
    # Converter de volta para texto
    texto_sem_stopwords = ' '.join(tokens_sem_stopwords)
    return texto_sem_stopwords
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_sem_stopwords'] = dataframe['text'].apply(stop_words)
# Filtrar as tokens que não são stop words
```

```python
dataframe.head(5)
```

Teste unitário da função


```python
def test():
  text = "A IBM é uma empresa de tecnologia muito antiga no mercado."

  rst = stop_words(text)

  if (rst == 'IBM empresa tecnologia antiga mercado .'):
    return True
  else:
    return False

test()
```

## 4.3 Lematização


Teste unitário da função


```python
# Definir a função que você deseja aplicar
def lematizar(valor):
    # Processamento de lematização usando o SpaCy
    doc = nlp(valor)
    # Extraia os lemas das palavras
    lemmas = [token.lemma_ for token in doc]
    # Converta os lemas em uma string
    lemmatized_text = " ".join(lemmas)
    return lemmatized_text
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_lemmas'] = dataframe['text_sem_stopwords'].apply(lematizar)
```

```python
dataframe.head(5)
```

Teste unitário da função


```python
def test():
  text = "A IBM é uma empresa de tecnologia muito antiga no mercado."

  rst = lematizar(text)

  if (rst == 'o IBM ser um empresa de tecnologia muito antigo em o mercado .'):
    return True
  else:
    return False

test()
```

## 4.4. Segmentação das palavras

### 4.4.1 Tokenização

```python
# Definir a função que você deseja aplicar
def tokenizar(valor):
    doc = nlp(valor)
    tokens = [token for token in doc]
    return tokens
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_tokens'] = dataframe['text_lemmas'].apply(tokenizar)
```

```python
dataframe.head(5)
```

Teste unitário da função


```python
def test():
  text = "A IBM é uma empresa de tecnologia muito antiga no mercado."

  rst = tokenizar(text)

  expected_tokens = ['A', 'IBM', 'é', 'uma', 'empresa', 'de', 'tecnologia', 'muito', 'antiga', 'no', 'mercado', '.']
  # Converta os objetos Token para strings e compare com os tokens esperados
  rst_text = [token.text for token in rst]

  if (rst_text == expected_tokens):
    return True
  else:
    return False

test()
```

### 4.4.3 Identificação das Entidades

```python
def entidades(valor):
    # Processar o texto
    doc = nlp(valor)
    # Iterar nas entidades previstas
    all_entidades = []
    for ent in doc.ents:
        # Imprimir o texto e a etiqueta da entidade
        #print(ent.text, ent.label_)
        all_entidades.append(ent.text)
    return all_entidades
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_entidades'] = dataframe['text_sem_stopwords'].apply(entidades)
```

```python
dataframe
```

Teste unitário da função


```python
def test():
  text = "A IBM é uma empresa de tecnologia muito antiga no mercado."

  rst = entidades(text)

  if (rst == ['IBM']):
    return True
  else:
    return False

test()
```

## 4.5 Pipeline de testes

```python
def pipeline(texto):

  texto = texto.lower()

  texto = stop_words(texto)

  texto = lematizar(texto)

  palavras = tokenizar(texto)

  return palavras
```
