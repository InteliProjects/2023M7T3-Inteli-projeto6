## 4.1. Setup do spaCy

```python
nlp = spacy.load("pt_core_news_lg")
nlp
```

## 4.2 Tirar as stop words

```python
# Definir a função que você deseja aplicar
def stop_words(valor):
    doc = nlp(valor)
    tokens_sem_stopwords = [token.text for token in doc if not token.is_stop]
    # Converter de volta para texto
    texto_sem_stopwords = ' '.join(tokens_sem_stopwords)
    return texto_sem_stopwords
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_tratado'] = dataframe['text'].apply(stop_words)
# Filtrar as tokens que não são stop words
```

```python
dataframe.head(5)
```

Teste unitário da função


```python
def test():
  text = "A IBM é uma empresa de tecnologia muito antiga no mercado."

  rst = stop_words(text)

  if (rst == 'IBM empresa tecnologia antiga mercado .'):
    return True
  else:
    return False

test()
```

## 4.3 Lematização


```python
# Definir a função que você deseja aplicar
def lematizar(valor):
    # Processamento de lematização usando o SpaCy
    doc = nlp(valor)
    # Extraia os lemas das palavras
    lemmas = [token.lemma_ for token in doc]
    # Converta os lemas em uma string
    lemmatized_text = " ".join(lemmas)
    return lemmatized_text
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_tratado'] = dataframe['text_tratado'].apply(lematizar)
```

```python
dataframe.head(5)
```

Teste unitário da função


```python
def test():
  text = "A IBM é uma empresa de tecnologia muito antiga no mercado."

  rst = lematizar(text)

  if (rst == 'o IBM ser um empresa de tecnologia muito antigo em o mercado .'):
    return True
  else:
    return False

test()
```

## 4.5 Regex - tirar números e caracteres especiais

```python
import re

def remover_numeros(valor):
    # Uso da expressão regular para encontrar e substituir todos os números pelo caractere vazio ('')
    texto_sem_numeros = re.sub(r'["”“._%$\d]', '', valor)
    return texto_sem_numeros
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_tratado'] = dataframe['text_tratado'].apply(remover_numeros)
```

```python
# Definir a função que você deseja aplicar
def tokenizar(valor):
    doc = nlp(valor)
    tokens = [token for token in doc]
    return tokens
```

### 4.4.3 Identificação das Entidades

```python
def entidades(valor):
    # Processar o texto
    doc = nlp(valor)
    # Iterar nas entidades previstas
    all_entidades = []
    for ent in doc.ents:
        # Imprimir o texto e a etiqueta da entidade
        #print(ent.text, ent.label_)
        all_entidades.append(ent.text)
    return all_entidades
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_entidades'] = dataframe['text_tratado'].apply(entidades)
```

# 5. Vetorização: Word Embedding CBOW

Será utilizado um modelo já treinado de CBOW, que utiliza redes neurais. ("CBOW 50 dimensões" do site do NILC (http://nilc.icmc.usp.br/nilc/index.php/repositorio-de-word-embeddings-do-nilc))

## 5.1 CBOW do NILC

```python
cbow = "/content/drive/Shareddrives/G6M7 Drive/cbow_s50.txt"
```

```python
from scipy.spatial.distance import cosine

from gensim.models import KeyedVectors

modelo_cbow = KeyedVectors.load_word2vec_format(cbow)
```

```python
def vetorizar(valor):
    lixos = [] #palavras que serão desconsideradas
    # Processar o texto
    doc = nlp(valor)
    all_vectors = []
    for vec in doc:
        #filtra palavras que não existem / erros de digitação
        if vec.text in modelo_cbow:
            wordvec = modelo_cbow[vec.text]
            all_vectors.append(wordvec)
        else:
            lixos.append(vec.text)
    #print(lixos)
    return all_vectors
```

# 7. Pipeline de testes

```python
def pipeline(texto):

  texto = lower_text(texto)

  texto = stop_words(texto)

  texto = lematizar(texto)

  palavras = tokenizar(texto)

  vetores = vetorizar(texto)

  return vetores
```
