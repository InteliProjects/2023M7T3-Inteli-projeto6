## 4.2 Tirar as stop words

### 4.2.1 Definição da função

```python
# Definir a função que você deseja aplicar
def stop_words(valor):
    doc = nlp(valor)
    tokens_sem_stopwords = [token.text for token in doc if not token.is_stop]
    # Converter de volta para texto
    texto_sem_stopwords = ' '.join(tokens_sem_stopwords)
    return texto_sem_stopwords
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_tratado'] = dataframe['text'].apply(stop_words)
```

```python
dataframe.head(5)
```

### 4.2.2 Teste unitário da função


```python
def test():
  text = "A IBM é uma empresa de tecnologia muito antiga no mercado."

  rst = stop_words(text)

  if (rst == 'IBM empresa tecnologia antiga mercado .'):
    return True
  else:
    return False

test()
```

## 4.3 Lematização


### 4.3.1 Definição da função

```python
# Definir a função que você deseja aplicar
def lematizar(valor):
    # Processamento de lematização
    doc = nlp(valor)
    # Extraia os lemas das palavras
    lemmas = [token.lemma_ for token in doc]
    # Converta os lemas em uma string
    lemmatized_text = " ".join(lemmas)
    return lemmatized_text
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_tratado'] = dataframe['text_tratado'].apply(lematizar)
```

```python
dataframe.head(5)
```

### 4.3.2 Teste unitário da função


```python
def test():
  text = "A IBM é uma empresa de tecnologia muito antiga no mercado."

  rst = lematizar(text)

  if (rst == 'o IBM ser um empresa de tecnologia muito antigo em o mercado .'):
    return True
  else:
    return False

test()
```

## 4.4 Tirar Uppercases


### 4.4.1 Definição da função

```python
# Definir a função que você deseja aplicar
def lower_text(valor):
    doc = nlp(valor)

    tokens = [token.text.lower() for token in doc]
    texto = ' '.join(tokens)
    return texto
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_tratado'] = dataframe['text_tratado'].apply(lower_text)
```

## 4.5 Regex - tirar números e caracteres especiais

### 4.5.1 Definição da função

```python
import re

def remover_numeros(valor):
    # Uso da expressão regular para encontrar e substituir todos os números pelo caractere vazio ('')
    texto_sem_numeros = re.sub(r'["”“._%$\d]', '', valor)
    texto_sem_numeros = re.sub(r'[0-9\[\]]', '', texto_sem_numeros)
    # texto_sem_espacos_duplos = re.sub(r'\s+', '', texto_sem_numeros) #remover espaços duplos
    return texto_sem_numeros
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_tratado'] = dataframe['text_tratado'].apply(remover_numeros)
```

```python
dataframe.head(5)
```

### 4.5.2 Teste unitário da função

```python
def test():
  text = "A empresa está a 100 anos no mercado."

  rst = remover_numeros(text)

  if (rst == 'A empresa está a  anos no mercado'):
    return True
  else:
    return False

test()
```

### 4.4.3 Identificação das Entidades

Definição da função

```python
def entidades(valor):
    # Processar o texto
    doc = nlp(valor)
    # Iterar nas entidades previstas
    all_entidades = []
    for ent in doc.ents:
        # Imprimir o texto e a etiqueta da entidade
        #print(ent.text, ent.label_)
        all_entidades.append(ent.text)
    return all_entidades
```

```python
# Aplicar a função à coluna específica em cada linha
dataframe['text_entidades'] = dataframe['text_tratado'].apply(entidades)
```

```python
dataframe.head(5)
```

Teste unitário da função

# 5. Vetorização: Word Embedding CBOW

Será utilizado um modelo já treinado de CBOW, que utiliza redes neurais. ("CBOW 50 dimensões" do site do NILC (http://nilc.icmc.usp.br/nilc/index.php/repositorio-de-word-embeddings-do-nilc))

## 5.1 CBOW do NILC

```python
cbow = "/content/drive/Shareddrives/G6M7 Drive/cbow_s50.txt"
```

```python
from scipy.spatial.distance import cosine

from gensim.models import KeyedVectors

modelo_cbow = KeyedVectors.load_word2vec_format(cbow)
```

### 5.1.1 Definição da função

```python
def vetorizar(valor):
    lixos = [] #palavras que serão desconsideradas
    # Processar o texto
    doc = nlp(valor)
    all_vectors = []
    for vec in doc:
        #filtra palavras que não existem / erros de digitação
        if vec.text in modelo_cbow:
            wordvec = modelo_cbow[vec.text]
            all_vectors.append(wordvec)
        else:
            lixos.append(vec.text)
    #print(lixos)
    return all_vectors
```

# 6. Cálculo da similaridade

Utilizaremos a métrica "similaridade do cosseno" para fazer a comparação entre os inputs do usuário e os documentos disponíveis.

```python
def similaridade(input,doc):
    result = 1 - cosine(input, doc)
    return result
```

# 7. Pipelines de testes

## 7.1 Pipeline de tratamento do input

### 7.1.1Definição da função

```python
def pipelineTratamento(input):

  texto = stop_words(input)

  texto = lematizar(texto)

  texto = lower_text(texto)

  texto = remover_numeros(texto)

  vetores = vetorizar(texto)

  vetor = vetorUnico(vetores)

  return vetor
```

### 7.1.2 Teste unitário da função

```python
texto = "O rato roeu a roupa do rato"

vetorFinal = pipelineTratamento(texto)

vetorFinal
```

## 7.2 Pipeline final do modelo

A função receberá como input uma frase do usuário, e terá como output as 5 maiores similaridades em relação à base de dados, e seus respectivos IDs, para que os itens possam ser acessados.

### 7.2.1 Definição da função

```python
import pandas as pd
import numpy as np
import json

def pipeline(input):
    tratado = pipelineTratamento(input)
    array=[]
    for item in dataframe['vectors']:
        sim = similaridade(tratado, item)
        array.append(sim)

    df = pd.DataFrame(array, columns=["vec"])

    k = 5

    maiores_valores = df["vec"].nlargest(k)
    indices_maiores = maiores_valores.index

    noticias = []
    for i in indices_maiores:
        noticias.append(dataframe['text'][i])

    noticias_dict = {
        "noticia1": noticias[0],
        "noticia2": noticias[1],
        "noticia3": noticias[2],
        "noticia4": noticias[3],
        "noticia5": noticias[4]
    }
    # json_data = json.dumps(noticias_dict)

    return noticias_dict
```

## 8. Testes adicionais

```python
def compararFrases(input1, input2):
    tratado1 = pipelineTratamento(input1)
    tratado2 = pipelineTratamento(input2)

    sim = similaridade(tratado1,tratado2)
    return sim
```
